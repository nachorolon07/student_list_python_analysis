GENDER DISTRIBUTION


```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("gender distribution")
plt.show()
```

```python
#from the above chart we can analysed that the no. of female in the data is more than the male
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Parent's education impact on student score")
plt.show()
```

```python
#from the above chart we have conclude that the education of the parents have a good impact on their students score
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("parent's maritial status impact on student's score")
plt.show()
```

```python
#from the above chart we can see that their is no/neglegible impact on student's
#score due to parent's maritial status
```

# **DISTRIBUTION OF ETHENIC GROUP**


```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

list1=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l=["groupA","groupB","groupC","groupD","groupE"]
plt.pie(list1,labels=l,autopct="%1.2f%%")
plt.title("distribution of ethenic groups")
plt.show()
```
